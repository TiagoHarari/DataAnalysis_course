d) ¿Cuál era la tasa de desocupación en el período seleccionado para las mujeres? ¿Era más alta o más baja que la de los hombres? (Extra: pensar el resultado usando una función)

Nota: la tasa se calcula como Desocupados / (Ocupados + Desocupados)

````python
# tasa de desocupacion = Desocupados / (Ocupados + Desocupados)
def obtener_tasa_desocupacion(df): # definimos una funcion
  eph_ocupados = df[df['estado'] == 1] # df de personas en estado de ocupacion
  eph_desocupados = df[df['estado'] == 2] #df de personas en estado de desocupacion
  total_ocupados = eph_ocupados['pondera'].sum() #total ponderado de ocupados
  total_desocupados = eph_desocupados['pondera'].sum() #total ponderado de desocupados

  return round(total_desocupados / (total_ocupados + total_desocupados)*100, 3) # tasa de desocupacion redondeada a 3 decimales

tasa_mujer = obtener_tasa_desocupacion(eph_mujeres)
tasa_hombre = obtener_tasa_desocupacion(eph_hombres)
print("La tasa de desocupacion de las mujeres es: ",tasa_mujer)
print("La tasa de desocupacion de los hombres es: ",tasa_hombre)

if tasa_mujer > tasa_hombre:
  print("La tasa de desocupacion de mujeres es mayor que la de hombres")
else:
  print("La tasa de desocupacion de los hombres es mayor que la de mujeres")
````
